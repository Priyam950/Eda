# DATA CLEANING

creating a copy of the data for manuplation or processing

```python
new_data= data.copy()
```

```python
new_data.head(5)
```

```python
# Changing the data types
new_data['SeniorCitizen'] = new_data['SeniorCitizen'].astype(str)

# Convert 'TotalCharges' to numeric, handling errors by coercing to NaN
new_data['TotalCharges'] = pd.to_numeric(new_data['TotalCharges'], errors='coerce')
```

```python
new_data.info()
```

as we see there are some missing values in TotalCharges after changing the data types

```python
#Let's check
new_data[new_data['TotalCharges'].isna()==1]
```

```python
new_data.isna().sum()/len(new_data)*100
```

The missing data is very less that'why we can simply remove than

```python
new_data.dropna(how='any', inplace=True)
```

Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on..

```python
print(new_data['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

new_data['tenure_group'] = pd.cut(data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
new_data['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure
new_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
new_data.head()
```

```python
# Set the style of seaborn for better visualization
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot KDE for MonthlyCharges with Churn separation
sns.kdeplot(data=new_data, x="MonthlyCharges", hue="Churn", fill=True, common_norm=False, palette="husl", alpha=.7, linewidth=0, ax=axes[0])
axes[0].set_title('KDE Plot - MonthlyCharges')
axes[0].set_xlabel('Monthly Charges')
axes[0].set_ylabel('Density')

# Plot KDE for TotalCharges with Churn separation
sns.kdeplot(data=new_data, x="TotalCharges", hue="Churn", fill=True, common_norm=False, palette="husl", alpha=.7, linewidth=0, ax=axes[1])
axes[1].set_title('KDE Plot - TotalCharges')
axes[1].set_xlabel('Total Charges')
axes[1].set_ylabel('Density')

plt.show()
```

Total charges increase as monthly charge increase

Churn is high when monthly charges is high
*Surprising* higly churn in less total charges
